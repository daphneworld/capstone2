```python
# importing relevant packages
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
```

```python
df=pd.read_csv('Auto_Insurance_Claims_Sample (1).csv')
print(df.head())
```

```python
# exploring the data
print(df.shape)
print(df.columns)
```

```python
print(df.dtypes)
print(df.describe())
```

```python
# Removing the unnecessary information columns
data=df.drop(columns=['Country'])
# print the new size of data
print(data.shape)
```

```python
data.State.value_counts()
```

```python
# Checking the duplicated rows
data.duplicated().value_counts()
```

```python
# Checking the missing values
data.isnull().sum()
```

```python
# Removing the white spaces in column names
data.columns = data.columns.str.replace(' ', '')
# print the columns names
print(data.columns)
```

```python
data.EffectiveToDate.str.replace('/', '-')
```

```python
# Changing data type
data['EffectiveToDate'] = pd.to_datetime(data['EffectiveToDate'], errors='coerce')
```

```python
# exporting data
data.to_csv('Auto_Insurance_Dataset.csv', index = False)
```
